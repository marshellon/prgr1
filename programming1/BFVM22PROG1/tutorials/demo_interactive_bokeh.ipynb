```python
import numpy as np
import pandas as pd
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.plotting import ColumnDataSource

import panel as pn

output_notebook()
pn.extension()
```

```python
data = np.random.randint(0, 100, 100)
df = pd.DataFrame({'data': data})
```

```python
df.info()
```

```python
df['smooth'] = df['data'].rolling(3).mean()
```

```python
def smooth(window):
    plotdata.data['smooth'] = df['data'].rolling(window).mean()
    pn.io.push_notebook(bokeh_pane)
```

```python
p = figure()
p.xaxis.axis_label = 'time'
p.yaxis.axis_label = 'signal'
plotdata = ColumnDataSource(df)
p.line('index', 'data', source=plotdata)
p.line('index', 'smooth', source=plotdata, color='red')

slider = pn.widgets.IntSlider(start=1, end=10, value=3, step=1)
pn.interact(smooth, window=slider)

bokeh_pane = pn.Column(pn.pane.Bokeh(p), slider)
bokeh_pane
```
